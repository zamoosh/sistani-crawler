```python
from bs4.element import Tag
from bs4 import BeautifulSoup
from requests import Response
import requests
import re

URL_REGEX = re.compile(r".*(\d+/\d+)/")
BASE_URL = "https://www.sistani.org"
URL_LIST = [
    "https://www.sistani.org/persian/book/26575/",
    "https://www.sistani.org/persian/book/26576/",
    "https://www.sistani.org/persian/book/26577/",
    "https://www.sistani.org/persian/book/26578/",
]
```

```python
res: Response = requests.get(URL_LIST[0])
```

```python
res_body = res.content.decode("UTF-8")
soup: BeautifulSoup = BeautifulSoup(res_body, "html.parser")

for i in soup.find_all("a"):  # type: Tag
    if i.get("href") is not None and len(URL_REGEX.findall(i.get("href"))) > 0:
        print(BASE_URL + i.get("href"))
```
